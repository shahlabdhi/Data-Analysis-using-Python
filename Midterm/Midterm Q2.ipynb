Storing data based on id

```python
import os.path
import glob
import json

pathname="Downloaded Data/Article Search Data"
newpath="Downloaded Data/Article Search Data Storage/"

uniqueids={}
my_dict ={}

for filename in glob.glob(os.path.join(pathname, '*.json')):
    with open(filename) as f:
        d = json.load(f)
        if 'response' in d:
            doc=d['response']['docs']

            for i in range(0,1500000):
                try:
                    id = doc[i]['_id']
                    result = []
                    if id not in uniqueids:
                        my_dict={}
                        uniqueids[id]=1
                        #print(id)
                        my_dict['_id']=id
                        my_dict['lead_paragraph']= doc[i]['lead_paragraph']
                        
                        my_dict['section_name']=doc[i]['section_name']
                        
                        my_dict['pub_date']=doc[i]['pub_date']
                       
                        my_dict['type_of_material']=doc[i]['type_of_material']
                        
                        my_dict['document_type']=doc[i]['document_type']
                       
                        my_dict['headline']=doc[0]['headline']
                        
                    
                        #my_dict['main']=doc[i]['headline'][0]['main']
                        
                        my_dict['source']=doc[i]['source']
                        
                        my_dict['snippet']=doc[i]['snippet']
                        
                        my_dict['web_url']=doc[i]['web_url']
                        
                        result.append(my_dict)
                      
                        with open(os.path.join(newpath+id+'.json'), 'w') as outfile:
                            json.dump(result,outfile)
                       
                            
                except(IndexError,KeyError):
                    continue
            
```

# Analysis on Data

# Find Web-url of all Articles which were based on Photography between 2015 to 2017 and store in a text file

```python
import os.path
import glob
import json

pathname="Downloaded Data/Archive Data"
newpath = "Downloaded Data/Archive Data/Processed Data"
with open(os.path.join(newpath,'webaddresses.txt'), 'w') as outfile:
    for filename in glob.glob(os.path.join(pathname, '*.json')):
        if(filename[39:-5]=='2015' or filename[39:-5]=='2016' or filename[39:-5]=='2017'):
            with open(filename) as f:
                d = json.load(f)
                if 'response' in d:
                    doc=d['response']['docs']
                    for i in range(0,1500000):    
                        try:
                            keyword=doc[i]['keywords'][0]
                            category=keyword['value']
                            if(category=='Photography'):
                                webaddress = doc[i]['web_url']
                                json.dump(webaddress,outfile)
                                outfile.write('\n')
                        except(IndexError,KeyError):
                            continue
```

# Generate a report to see various sources of articles

```python
import os.path
import glob
import json
import csv

pathname="Downloaded Data/Article Search Data Storage/"
mydict={}
mydict['Times']=0
mydict['AP']=0
mydict['Reuters']=0
mydict['Others']=0

for filename in glob.glob(os.path.join(pathname, '*.json')):
    with open(filename) as f:
        d = json.load(f)
        source=d[0]['source']
        
        try:
            if(source=='The New York Times'):
                #print(source)
                mydict['Times']+=1
                #print(mydict['Times'])
            elif (source=="AP"):
                mydict['AP']+=1
            elif(source=="Reuters"):
                mydict['Reuters']+=1
            else:           
                mydict['Others']+=1
        except(KeyError):
            print("I am here")
            continue
with open("SourceReport.csv", "w",encoding='utf-8') as fp:
    header=['New York Times Count','AP Count','Reuters Count','Others Count']
    writer = csv.DictWriter(fp, fieldnames=header,lineterminator='\n')
    writer.writeheader()
    writer.writerow({'New York Times Count': mydict['Times'],
                     'AP Count': mydict['AP'],
                     'Reuters Count': mydict['Reuters'],
                     'Others Count': mydict['Others']
                    })
                    
```

# Create folders based on different types of articles - Review, Sports, News etc and then determine the highest numbers of articles belong to which type

```python
import os.path
import glob
import json


pathname="Downloaded Data/Article Search Data Storage/"
newpath ="Downloaded Data/Data Processed/"

for filename in glob.glob(os.path.join(pathname, '*.json')):
    with open(filename) as f:
        d = json.load(f)
        categories=d[0]['type_of_material']
        if os.path.isdir(newpath+categories):
            print("Already exists")
        else:
            os.makedirs(newpath+categories)
        newpath=newpath+categories
        filenam = os.path.basename(filename)
        with open(os.path.join(newpath,filenam), 'w') as outfile:
            json.dump(d, outfile)
        newpath ="Downloaded Data/Data Processed/"
max=1
for f in os.listdir(newpath):
    path1=newpath+f
    for filename in glob.glob(path1):
        num_files = len([fname for fname in os.listdir(path1)
                if os.path.isfile(os.path.join(path1, fname))])
        print("Number of files present in "+f+" is:",num_files)    
        if(max<num_files):
            max=num_files
            maxname = f
print("Maximum type of articles are present in "+maxname+" which are:",max)
        
```

# Most popular words used in 2015

```python
import os.path
import glob
import json
from collections import Counter
import csv

pathname="Downloaded Data/Archive Data"
wordcount={}

for filename in glob.glob(os.path.join(pathname, '*.json')):
    if(filename[39:-5]=='2015'):
    #if(filename[44:-5]=='2015'):
        with open(filename) as f:
            d = json.load(f)
            if 'response' in d:
                doc=d['response']['docs']
                for i in range(0,1500000):
                    try:
                        para = doc[i]['lead_paragraph']
                        #print(para)
                        if para is not None:
                            for word in para.split():
                                #print(word)
                                word = word.lower()
                                if word not in wordcount:
                                    wordcount[word] = 1
                                else:
                                    wordcount[word] += 1
                    except(IndexError,KeyError):
                        continue
                        
c=Counter(wordcount)
print(c.most_common()[:20])
    
```
