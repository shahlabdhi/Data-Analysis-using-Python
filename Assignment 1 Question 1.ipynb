# Question 1
- Files present in Gutenberg
- Words and Unique Words count in 'austen-persuasion.txt'
- Frequency for each word and list top 10 most frequent words with their frequency

```python
import nltk
nltk.corpus.gutenberg.fileids()
```

```python
wordcount = nltk.corpus.gutenberg.words('austen-persuasion.txt')
print("Words present in austen-persuasion are", len(wordcount))
print("Unique Words present in austen-persuasion are", len(set(wordcount)))
```

```python
file= open("C:/nltk_data/corpora/gutenberg/austen-persuasion.txt")
wordcount={}
for word in file.read().split():
    if word not in wordcount:
        wordcount[word] = 1
    else:
        wordcount[word] += 1
for k,v in wordcount.items():
    print(k, v)
```

```python
file= open("C:/nltk_data/corpora/gutenberg/austen-persuasion.txt")
wordcount={}
for word in file.read().split():
    if word not in wordcount:
        wordcount[word] = 1
    else:
        wordcount[word] += 1

copy = []
for k,v in wordcount.items():
    copy.append((v, k))


copy = sorted(copy, reverse=True)

for k in copy[:10]:
    print(k[1], k[0])
```
