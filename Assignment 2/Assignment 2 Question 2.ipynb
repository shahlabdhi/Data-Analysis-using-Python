```python
import os.path
import glob
import requests
import json

def path():
     return "D:/Python/DataAnalysis4Python_Spring17-master/Assignment 2/Data"

pathname = path()
newpath = 'D:/Python/Data Processed/' 
countname = {}

for filename in glob.glob(os.path.join(pathname, '*.json')):
    with open(filename) as f:
        d = json.load(f)
        countryname = d['location']['country']
        cityname = d['location']['city']
        if os.path.isdir(newpath+countryname):
            print("Already exists")
        else:
            os.makedirs(newpath+countryname)
            
        if os.path.isdir(newpath+countryname+'/'+cityname):
            print("This city folder already exists")
        else:
            os.makedirs(newpath+countryname+'/'+cityname)
         
        filenam = os.path.basename(filename)
        newpath = newpath+countryname+'/'+cityname
        with open(os.path.join(newpath,filenam), 'w') as outfile:
            json.dump(d, outfile)
            
        newpath ='D:/Python/Data Processed/'
```
