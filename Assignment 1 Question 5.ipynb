# Questions
- Top 5 words that Shakespeare used but we do not use currently
- Top 50 words from Shakespeare (all 3 books) and top 50 from Web_text 
- Remove punctuation and stop words
- Remove the words we still use today, and get the unused list. Show the top 5 elements

```python
import nltk
L=[]
L = nltk.corpus.gutenberg.fileids()
words_txt = []
for name in L:
    if name.find('shakespeare')!=-1:
        #print(name)
        words_txt = nltk.corpus.gutenberg.words(name)
        words_txt = [x.lower() for x in words_txt]
        wdset = (words_txt)
M = []
M = nltk.corpus.webtext.fileids()
for name in M:
        words_txt1 = nltk.corpus.webtext.words(name)
        words_txt1 = [x.lower() for x in words_txt1]
        wdset1 = set(words_txt1)

oldwords = {}
count=0
for y in wdset:
    if y not in wdset1:
        if y not in oldwords:
            oldwords[y]=1
        oldwords[y]+=1
#print(oldwords)
copy = []
for k,v in oldwords.items():
    copy.append((v, k))


copy = sorted(copy, reverse=True)

for k in copy[:5]:
    print("Top 5 words that Shakespeare used but are not used now are:",k[1], k[0])
```
